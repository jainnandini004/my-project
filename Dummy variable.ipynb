# Dummy Variable

```python
import pandas
```

```python
thd=pandas.read_csv("Transformed_Housing_Data.csv")
```

```python
thd=thd.rename(columns={"Condition of the House":"Condition_of_the_House"})
```

```python
thd.head()
```

```python
thd=pandas.get_dummies(thd,columns=["Condition_of_the_House"],drop_first=True)
```

```python
thd.head()
```

```python
thd=pandas.get_dummies(thd,columns=["Ever Renovated","Waterfront View"])
```

```python
thd.head()
```

# Binning

```python
zip_Table=thd.groupby("Zipcode").agg({"Sale_Price":"mean"}).sort_values("Sale_Price",ascending=True)
zip_Table
```

```python
zip_Table["Zipcode_Group"]=pandas.cut(zip_Table["Sale_Price"],bins=10,labels=["zip_group_1","zip_group_2","zip_group_3","zip_group_4","zip_group_5","zip_group_6","zip_group_7","zip_group_8","zip_group_9","zip_group_10"],include_lowest=True)
```

```python
zip_Table=zip_Table.drop(columns="Sale_Price")
```

```python
thd=pandas.merge(thd,zip_Table,left_on="Zipcode",how="left",right_index=True)
thd=thd.drop(columns="Zipcode")
thd
```
