# Multicollinearity

```python
#After preprocessing
```

```python
import pandas
data=pandas.read_csv("Transformed_Housing_Data2(1).csv")
data.head()
```

```python
#Scaling
```

```python
from sklearn.preprocessing import *
scaler=StandardScaler()
Y=data["Sale_Price"]
X=scaler.fit_transform(data.drop(columns=["Sale_Price"]))
X=pandas.DataFrame(data=X,columns=data.drop(columns=["Sale_Price"]).columns)
X.head()
```

```python
#check correlation
```

```python
X.corr()
```

```python
z=[]
k=X.corr()
for i in k.columns:
    for j in k.columns:
        if (k.loc[i,j]>abs(0.5))&(i!=j):
            z.append([str(i),str(j)])
z,len(z)            
            
```

# Calculating VIF

```python
import pandas as pd
from statsmodels.stats.outliers_influence import *
vif_data=X
VIF=pd.Series([variance_inflation_factor(vif_data.values,i) for i in range(vif_data.shape[1])],index=vif_data.columns)
VIF,len(vif_data)
```

```python
VIF[VIF==VIF.max()].index[0]
```

```python
def MC_remove(data):
    VIF=pd.Series([variance_inflation_factor(vif_data.values,i) for i in range(vif_data.shape[1])],index=vif_data.columns)
    if VIF.max()>5:
        print(VIF[VIF==VIF.max()].index[0],"has been removed")
        data=data.drop(columns=VIF[VIF==VIF.max()].index[0])
        return data
    else:
        print("No Multicollinearity present anymore")
        return data
for i in range(7):
    vif_data=MC_remove(vif_data)

vif_data.head()    
    
```

# Calculating vif for remaining columns

```python
VIF=pd.Series([variance_inflation_factor(vif_data.values,i) for i in range(vif_data.shape[1])],index=vif_data.columns)
VIF,len(vif_data.columns)
```

# Train/Test set

```python
 x=vif_data
 y=data["Sale_Price"]   
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=101)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

# Implement linear regression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression(normalize=True)
lr.fit(x_train,y_train) 
```

```python
lr.coef_  #BEST VALUE OF M1,m2.....to avoid long procedure
```

```python
#R^2
predictions=lr.predict(x_test)
lr.score(x_test,y_test)# calculate R^2 for performance of modelx 
```

```python

#adj R^2
def adj_r2(data_x,data_y,predictions):#data_x=data contain independent variable and data_y=data contain label prediction
    from sklearn.metrics import r2_score
    R=r2_score(data_y,predictions)
    n=len(data_y)
    m=len(data_x.columns)
    adj_R=1-((1-R)*(n-1))/(n-m-1)
    return adj_R
```

```python
adj_r2(x_test,y_test,predictions)
```

# Verifying the assumptions

```python
#1.RESIDUALS
```

```python
residuals=predictions-y_test
residual_table=pd.DataFrame({"residuals":residuals,"predictions":predictions})
residual_table=residual_table.sort_values(by="predictions")
residual_table.head()
```

```python
z=[i for i in range(int(residual_table["predictions"].max()))]
k=[0 for i in range(int(residual_table["predictions"].max()))]
```

```python
import matplotlib.pyplot as plt
plt.figure(dpi=130,figsize=(17,7))
plt.scatter(residual_table["predictions"],residual_table["residuals"],color="red",s=2)
plt.plot(z,k,color="green",linewidth=3,label="regression line")
 
plt.xlabel("fitted points(ordered by predictions)")
plt.ylabel("residuals")
plt.title("Residual Plot")
plt.legend()
plt.show()
```

```python
#DISTRIBUTION OF ERRORS
```

```python
plt.figure(dpi=100,figsize=(10,7))
plt.hist(residual_table["residuals"],color="red",bins=200)
plt.xlabel("RESIDUALS")
plt.ylabel("FREQUENCY")
plt.title("Distribution Of Residuals")
plt.show()#follow normal distribution bt some outlier are still present
```
