# Variable Transformation

```python
import pandas as pd
```

```python
rhd=pandas.read_csv("1. Regression - Module - (Housing Prices).csv")
```

```python
rhd
```

```python
import numpy as np
```

```python
rhd.dtypes
```

```python
rhd["No of Times Visited"].unique()
```

```python
mapping={"None":"0","Once":"1","Twice":"2","Thrice":"3","Four":"4"}
```

```python
rhd["No of Times Visited"]=rhd["No of Times Visited"].map(mapping)
```

```python
rhd["No of Times Visited"].unique()
```

```python
rhd
```

```python
rhd["Ever Renovated"]=np.where(rhd["Renovated Year"]==0,"No","Yes")
```

```python
rhd
```

```python
rhd["Purchase Year"]=pd.DatetimeIndex(rhd["Date House was Sold"]).year
```

```python
rhd["Year Since Renovation"]=np.where(rhd["Ever Renovated"]=="Yes",abs(rhd["Purchase Year"]-rhd["Renovated Year"]),0)
```

```python
rhd
```

```python
rhd.drop(columns=["Purchase Year","Date House was Sold","Renovated Year"],inplace=True)
```
