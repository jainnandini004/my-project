```python
#heare we have a lot of string data which we need to convert
sex=pd.get_dummies(titanic_data["Sex"],drop_first=True)
sex.head(5)
```

```python
titanic_data.head(5)
```

```python
enb=pd.get_dummies(titanic_data["Embarked"],drop_first=True)
```

```python
pclass=pd.get_dummies(titanic_data["Pclass"],drop_first=True)
```

```python
titanic_data=pd.concat([titanic_data,enb,pclass,sex],axis=1)
titanic_data.drop(columns=["Sex","Pclass","Embarked","Name","Ticket","PassengerId"],inplace=True)
```

```python
titanic_data.drop(columns=["Name","Ticket","PassengerId"],inplace=True)
```

```python
titanic_data.head(5)
```

# Train data


```python
#here we train by survive data
```

```python
X=titanic_data.drop(columns="Survived",axis=1)
Y=titanic_data["Survived"]#dependent variable
```

```python
from sklearn.model_selection import train_test_split
```

```python
#split data into test and train
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel=LogisticRegression()
logmodel.fit(X_train,Y_train)
# here we jut create a model fit the model and make predictions
```

```python
predictions=logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
k=classification_report(Y_test,predictions)
print(k)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(Y_test,predictions)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,predictions)
```
