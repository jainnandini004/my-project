```python
import pandas
```

```python
thd=pandas.read_csv("Transformed_Housing_Data.csv")
```

```python
thd
```

```python
from statsmodels.formula.api import ols
import statsmodels.api as sm
```

```python
thd=thd.rename(columns={"Sale Price":"Sale_Price"})
```

```python
thd=thd.rename(columns={"Condition of the House":"Condition_of_the_House"})
```

```python
mod=ols("Sale_Price~Condition_of_the_House",data=thd).fit()
```

```python
anova_table=sm.stats.anova_lm(mod,typ=2)
```

```python
anova_table
```

```python
thd=thd.rename(columns={"Ever Renovated":"Ever_Renovated"})
```

```python
thd=thd.rename(columns={"Waterfront View":"Waterfront_View"})
```

```python
mod=ols("Sale_Price~Waterfront_View",data=thd).fit()
```

```python
anova_table=sm.stats.anova_lm(mod,typ=2)
```

```python
anova_table
```
